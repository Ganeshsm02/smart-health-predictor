```python
import pandas as pd
diabetes = pd.read_csv("diabetes.csv")
heart = pd.read_csv("heart_statlog_cleveland_hungary_final.csv")
liver = pd.read_csv("indian_liver_patient.csv")
kidney = pd.read_csv("kidney_disease.csv")
print(diabetes.head())
print(heart.head())
print(liver.head())
print(kidney.head())
```

```python
diabetes["disease"] = diabetes["Outcome"].replace({1: "Diabetes", 0: "No Diabetes"})
heart["disease"] = heart["target"].replace({1: "Heart Disease", 0: "No Heart Disease"})
kidney["disease"] = kidney["classification"].replace({"ckd": "CKD", "notckd": "No CKD"})
liver["disease"] = liver["Dataset"].replace({1: "Liver Disease", 2: "No Liver Disease"})
diabetes.drop(['Outcome'],inplace=True,axis=1)
heart.drop(['target'],axis=1)
kidney.drop(['classification'],axis=1)
liver.drop(['Dataset'],axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
db = diabetes['disease']
diabetes['disease_int'] = db.replace({'Diabetes':1,'No Diabetes':0})
diabetes.drop(['disease'],axis=1,inplace = True)
```

```python
print(diabetes['disease_int'].value_counts())
```

```python
X = diabetes.drop(['disease_int'],axis=1)
Y = diabetes['disease_int']

```

```python
print(len(X.columns))
```

```python
diabetes.head()
```

```python
diabetes.fillna(method='bfill',inplace=True)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42,stratify=Y)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)
```

```python
diabetes_scaler = StandardScaler()
x_train_scale = diabetes_scaler.fit_transform(x_train_resampled)
x_test_scale = diabetes_scaler.transform(x_test)
```

```python
model = RandomForestClassifier(n_estimators=450,max_depth=5,random_state=42,class_weight='balanced')
model.fit(x_train_scale,y_train_resampled)
y_pred = model.predict(x_test_scale)
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python

import numpy as np
sample_input_low = np.array([[1, 95, 70, 20, 80, 24.5, 0.2, 25]])   
sample_input_med = np.array([[3, 130, 82, 30, 180, 32.5, 0.45, 40]])  
sample_input_high = np.array([[6, 180, 95, 40, 350, 38.7, 1.2, 55]])  


sample = sample_input_med

scaled_sample = diabetes_scaler.transform(sample)


prediction = model.predict(scaled_sample)
probability = model.predict_proba(scaled_sample)[0][1]

print("DIABETES MODEL TEST")
print("Prediction:", prediction)
print("Probability:", probability)


if probability < 0.4:
    print("Risk Level: LOW")
elif probability < 0.75:
    print("Risk Level: MEDIUM")
else:
    print("Risk Level: HIGH")
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report, roc_auc_score
import numpy as np


cv_scores = cross_val_score(model, x_train_resampled, y_train_resampled, cv=5, scoring='accuracy')
print("Cross-validation Accuracy scores:", cv_scores)
print("Mean CV Accuracy:", np.mean(cv_scores))

y_pred = model.predict(x_train_resampled)

print("TRAIN SET METRICS")
print("Accuracy:", accuracy_score(y_train_resampled, y_pred))
print("Precision:", precision_score(y_train_resampled, y_pred, average='weighted'))
print("Recall:", recall_score(y_train_resampled, y_pred, average='weighted'))
print("F1-score:", f1_score(y_train_resampled, y_pred, average='weighted'))
print("Confusion Matrix:\n", confusion_matrix(y_train_resampled, y_pred))
print("\n")


y_test_pred = model.predict(x_test_scale)

print("TEST SET METRICS")
print("Accuracy:", accuracy_score(y_test, y_test_pred))
print("Precision:", precision_score(y_test, y_test_pred, average='weighted'))
print("Recall:", recall_score(y_test, y_test_pred, average='weighted'))
print("F1-score:", f1_score(y_test, y_test_pred, average='weighted'))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred))
print("\nClassification Report:\n", classification_report(y_test, y_test_pred))


if len(set(y_test)) == 2:
    y_test_prob = model.predict_proba(x_test_scale)[:, 1]
    auc = roc_auc_score(y_test, y_test_prob)
    print("ROC-AUC:", auc)
```
